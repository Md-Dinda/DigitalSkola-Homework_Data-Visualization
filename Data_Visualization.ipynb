##Visualization of Columns in the Dataset

**Attrition_flag**

```python
df['Attrition_Flag'].unique()
```

```python
explode = (0.05, 0.05)
plt.pie(df['Attrition_Flag'].value_counts(), labels=['Existing Customer', 'Attrited Customer'], autopct='%1.1f%%', pctdistance=0.8, explode=explode)
centre_circle = plt.Circle((0, 0), 0.50, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Proportion of Active vs. Churned Customers')
plt.show()
```

```python
##Interpretation:
  #The company has a good customer retention rate, with the majority of customers (83.9%) still active.
  #However, there is still a significant proportion of customers (16.1%) who have churned.

##Based on the findings of the analysis, the company can take steps to address the reasons for churn and improve customer retention, such as:
  #Improving customer service and support.
  #Offering more competitive products or pricing.
  #Developing targeted marketing campaigns to re-engage churned customers.
```

## Visualization Using Countplot

```python
# Group data by Attrition_Flag
sns.countplot(
    x="Income_Category",
    y=None,
    hue="Attrition_Flag", #by Attrition_Flag
    data=df
)

plt.title("Distribution of Active and Exited Customers by Income Category")
plt.xlabel("Income Category")
plt.ylabel("Number of Customers")
plt.legend(title="Attrition Flag")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
##Notes:
#1# For all income categories, the number of active customers is significantly higher than the number of exited customers,
  # suggesting that the majority of customers remain loyal to their credit card service.
  #The ratio of exited to active customers is relatively high for the "Less than $40K" and "Unknown" categories, which may indicate that these customers are more likely to churn.
##Consider:
#1# Further analysis could be conducted to investigate the reasons behind the higher churn rate in these income categories.
  # This could involve examining other factors such as customer demographics, spending habits, and satisfaction with the credit card service.
#2# The company could consider implementing targeted strategies to reduce churn among customers in these income categories.
  # This could include offering incentives, such as rewards or lower interest rates, to encourage these customers to remain active.
```

## Visualization Using Scatterplot

```python
plt.scatter(df["Customer_Age"], df["Total_Trans_Amt"])

plt.title("Relationship between Customer Age and Number of Transactions")
plt.xlabel("Customer Age")
plt.ylabel("Number of Transactions")
plt.grid(True)
plt.show()
```

```python
## Notes:
#1# There are three main clusters in the distribution of transaction amounts: around 2,500 - 5,000, 7,500 - 10,000, and 12,500 - 15,000 transactions.
  # Within each age group (30-70 years old), there are customers with varying transaction amounts.
#2# Overall, this scatterplot suggests that there is no clear correlation between age and transaction amount. Customers of different ages tend to make similar numbers of transactions.
  # Since it is known that there is no strong correlation between age and transaction amount, the company can focus on retention strategies and offers that are not dependent on age.
#3# Additionally, promotions can be focused on the group with transaction amounts of 2,500 - 5,000 to encourage them to spend more, as this group represents the largest number of customers.
```

## Visualization Using HeatMap

```python
contingency_table = pd.crosstab(df['Income_Category'], df['Card_Category'])  # Contingency table
sns.heatmap(contingency_table, annot=True, fmt='d')

# Judul dan Label (Bahasa Indonesia)
plt.title("Hubungan antara Jenis Kartu dan Kategori Pendapatan")
plt.xlabel("Jenis Kartu")
plt.ylabel("Kategori Pendapatan")
plt.xticks(rotation=45, ha='right')

# Show Plot
plt.show()
```

```python
### Heatmap visualization is useful for seeing the relationship or pattern between two categorical variables and is better for large categories.
  # Therefore, as an example, the data in the 'Card_Category' and 'Income_Category' columns is chosen here.
## Keterangan:
## Notes:
#1# This heatmap is useful for identifying patterns or trends in the data, such as which card types are preferred within certain income ranges.
  # It uses darker colors to indicate higher numbers of cardholders.
  # The Blue card type is the most common across all income categories, with a higher proportion in the lower income categories.
  # As income category increases, the proportion of Platinum and Gold cardholders increases, while the proportion of Silver and Blue cardholders decreases.
  # This information could be used to tailor marketing campaigns or product offerings to specific income groups.
  # This can be a barrier for customers who are not willing to pay an additional fee for a credit card.
  # Additionally, the higher minimum monthly deposit requirement compared to other cards means that only customers with a certain income level will qualify for the Platinum card.
#2# A recommendation to retain Blue cardholders is to ensure customer satisfaction and engagement, especially for customers in the "80K - 120K" and "40K - 60K" categories,
  # which have a significant number of active customers..
  # To increase the number of higher-tier cardholders such as Gold and Platinum, the company could try to appeal to customers in the \\$120K+ category. Exclusive strategies such as premium services, personal financial consultants,
  # could be considered to attract more customers in this category.
```
